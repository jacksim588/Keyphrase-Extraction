```python
from rake_nltk import Rake
import nltk
r = Rake()
```

```python
def get_num_words(row):
    return len(str(row['Phrase']).split())

def get_perc_phrase(row,total):
    return (float(row['Score'])/total*100)
```

```python
def get_Rake_phrases(df):
    r = Rake()
    r.extract_keywords_from_sentences(df['text'])
    phrases=r.get_ranked_phrases_with_scores()
    df = pd.DataFrame(phrases, columns = ['Rank', 'Phrase'])
    df['Phrase'] = df['Phrase'].str.strip()
    phrase_counts = df['Phrase'].value_counts()
    phrase_counts = phrase_counts.reset_index()
    phrase_counts.columns=['Phrase','Score']
    phrase_counts['Length'] = phrase_counts.apply (lambda row: get_num_words(row), axis=1)
    phrase_counts=phrase_counts.loc[phrase_counts['Length'] <=6]
    total = float(phrase_counts['Score'].sum())
    phrase_counts['Perc'] = phrase_counts.apply (lambda row: get_perc_phrase(row,total), axis=1)
    return phrase_counts
```

```python
def calculate_frequency_change(current_year_df, previous_year_df, phrase_column='Phrase', frequency_column='Percentage Count'):
    """
    Calculates the change in frequency for phrases between two DataFrames, considering both
    increases and decreases in frequency, as well as phrases exclusive to each DataFrame.
    Assigns 100 to new phrases in the current year and -100 to phrases only found in the previous year.
    Keeps the phrase from the previous year.

    :param current_year_df: DataFrame for the current year.
    :param previous_year_df: DataFrame for the previous year.
    :param phrase_column: The name of the column containing the phrases.
    :param frequency_column: The name of the column containing the frequency percentages.
    :return: DataFrame with phrases and their change in frequency, including all other columns.
    """

    # Convert the phrase column to lowercase for consistent comparison
    current_year_df[phrase_column] = current_year_df[phrase_column].str.lower()
    previous_year_df[phrase_column] = previous_year_df[phrase_column].str.lower()

    # Merge the two DataFrames on the phrase column
    merged_df = pd.merge(current_year_df, previous_year_df, on=phrase_column, how='outer', suffixes=('_current', '_previous'))

    # Fill NaN values for the frequency columns
    merged_df[f"{frequency_column}_current"].fillna(0, inplace=True)
    merged_df[f"{frequency_column}_previous"].fillna(0, inplace=True)

    # Calculate the change in frequency
    # If the phrase is only in the current year, the change should be 100.
    # If the phrase is only in the previous year, the change should be -100.
    # Otherwise, it's the difference between the current and previous frequency.
    merged_df['frequency_change'] = merged_df.apply(
        lambda row: 100 if pd.isnull(row[f"{frequency_column}_previous"]) else
                    (-100 if pd.isnull(row[f"{frequency_column}_current"]) else
                     row[f"{frequency_column}_current"] - row[f"{frequency_column}_previous"]),
        axis=1
    )

    # Sort by the absolute change in frequency to see the most significant changes
    merged_df.sort_values(by='frequency_change', ascending=False, inplace=True)

    return merged_df
```

```python
df_2022 = pd.read_csv(r'Data/Phrases 2022.csv')
```

```python
df_2023 = pd.read_csv(r'Data/Phrases 2023.csv')
```

```python
frequency_df = calculate_frequency_change(current_year_df=df_2023, previous_year_df=df_2022)
```
