```python
import pandas as pd
import re
import csv
```

```python
brandbankExtract = r'Data/Bulk Extract Full.csv'
df = pd.read_csv(brandbankExtract)
```

```python
ingredients_toFind=[]
with open(r'F:\temp\Dairy ingredients.csv', 'r', encoding='utf-8-sig') as file:
    reader = csv.reader(file)
    for row in reader:
        string = row[0]
        ingredients_toFind.append(string.lower())
```

```python
ingredients_toFind
```

```python
# Create an empty list to store the matched rows
matched_rows = []

#drop blank ingredients
df = df.dropna(subset=['ingredients'])

# Iterate over each row in the dataframe
for index, row in df.iterrows():
    ingredients = row['ingredients']
    ingredients = ingredients.lower()
    for ingredient in ingredients_toFind:
        if re.search(r'\b' + re.escape(ingredient) + r'\b', ingredients):
            matched_rows.append({
                'name': row['name'],
                'gtin': row['gtin'],
                'pl1': row['pl1'],
                'pl2': row['pl2'],
                'pl3': row['pl3'],
                'ingredients': ingredients,
                'matched_ingredient': ingredient
            })

# Create a new dataframe from the matched rows
output_df = pd.DataFrame(matched_rows)

# Print the output dataframe
output_df
```

```python
output_df.to_csv('Data/Output/Dairy ingredients Extract.csv',index=False)
```
