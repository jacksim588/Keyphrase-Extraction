```python
df=df.fillna('')
#df = df[df['isDelete'] != 1]
df['text'] = df[[
    'description',
    'allergyAdvice',
    'recycling',
    'recycling_other',
    'brands',
    'marketing',
    'ingredients',
    'features',
    'storage',
    'preserves']].agg(' '.join, axis=1)

df.loc[df['isDelete'] == 1, 'text'] = ''


df['text'] = df['text'].str.lower()

#df['text'] = df['text'].str.replace('\n', '', regex=True)

df
```

```python
import pandas as pd
import numpy as np
dataset=df
# Assuming 'dataset' is your DataFrame
# Convert the 'date' column to datetime format, assuming dates are in 'yyyy-mm-dd' format
dataset['date'] = pd.to_datetime(dataset['date'], format='%Y-%m-%d', errors='coerce')

# Extract year and month
dataset['year'] = dataset['date'].dt.year
dataset['month'] = dataset['date'].dt.month

# Calculate word count for non-empty 'text' values
dataset['Word Count'] = dataset['text'].apply(lambda x: len(str(x).split()) if pd.notna(x) and str(x).strip() != "" else np.nan)

# Calculate average word count per month, ignoring empty or NaN values
average_word_count = dataset.groupby(['year', 'month'])['Word Count'].mean().reset_index()
average_word_count.rename(columns={'Word Count': 'Average Word Count'}, inplace=True)

# Count number of rows per month
monthly_count = dataset.groupby(['year', 'month']).size().reset_index(name='Monthly Count')
monthly_count['Monthly Cumulative Count'] = monthly_count['Monthly Count'].cumsum()

# Merge the two DataFrames on 'year' and 'month'
result_df = pd.merge(monthly_count, average_word_count, on=['year', 'month'])

# Save the combined DataFrame to a CSV file
result_df.to_csv(r'Data/Monthly Product Count.csv', index=False)
result_df
```

```python
from rake_nltk import Rake
import nltk
r = Rake()
```

```python
def get_num_words(row):
    return len(str(row['Phrase']).split())

def get_perc_phrase(row,total):
    return (float(row['Score'])/total*100)
```

```python
def get_Rake_phrases(df):
    r = Rake()
    r.extract_keywords_from_sentences(df['text'])
    phrases=r.get_ranked_phrases_with_scores()
    df = pd.DataFrame(phrases, columns = ['Rank', 'Phrase'])
    df['Phrase'] = df['Phrase'].str.strip()
    phrase_counts = df['Phrase'].value_counts()
    phrase_counts = phrase_counts.reset_index()
    phrase_counts.columns=['Phrase','Score']
    phrase_counts['Length'] = phrase_counts.apply (lambda row: get_num_words(row), axis=1)
    phrase_counts=phrase_counts.loc[phrase_counts['Length'] <=6]
    total = float(phrase_counts['Score'].sum())
    phrase_counts['Perc'] = phrase_counts.apply (lambda row: get_perc_phrase(row,total), axis=1)
    return phrase_counts
```

```python
df_raked = get_Rake_phrases(df)
#phrase_counts.to_csv(r'C:\Users\Clamfighter\Documents\GitHub\NLP\Impact Score Key Phrase\Extracted Phrases\Rake NLTK\Brandbank Latest Phrase Extraction.csv',index = False)
```
