```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
import torch
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
import nltk

# Download NLTK punkt tokenizer
nltk.download('punkt')
from nltk.tokenize import sent_tokenize

# Check if CUDA is available and set device accordingly
device = 0 if torch.cuda.is_available() else -1

# Load the tokenizer and model for BART-large MNLI (more accurate model)
tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-mnli")
model = AutoModelForSequenceClassification.from_pretrained("facebook/bart-large-mnli")
classifier = pipeline("zero-shot-classification", model=model, tokenizer=tokenizer, device=device)

# Extract label names for use in the DataFrame
label_names = [label.split(":")[0] for label in labels]

# Function to emphasize important keywords
def emphasize_keywords(text, keywords):
    for keyword in keywords:
        text = text.replace(keyword, f"{keyword} {keyword}")  # Repeat the keyword to emphasize
    return text

# Custom Dataset for DataLoader
class TextDataset(Dataset):
    def __init__(self, texts):
        self.texts = texts

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        return self.texts[idx]

# Function to classify claims in batches with dynamic threshold calculation
def classify_claims(input_data, batch_size=8):
    results = []
    thresholds = {}

    # If input_data is a single text string
    if isinstance(input_data, str):
        input_data = pd.DataFrame({'marketing': [input_data], 'name': ['sample_text']})

    # Filter rows with non-empty 'marketing' text
    input_data = input_data[input_data['marketing'].notna() & (input_data['marketing'] != "")]

    # Emphasize keywords
    input_data['marketing'] = input_data['marketing'].apply(lambda x: emphasize_keywords(x, keywords))

    # Convert 'marketing' texts to a DataLoader
    texts = input_data['marketing'].tolist()
    xml_ids = input_data['xml_id'].tolist()

    for xml_id, text in tqdm(zip(xml_ids, texts), total=len(texts), desc="Processing Rows"):
        sentences = sent_tokenize(text)
        for sentence in sentences:
            result = classifier(sentence, candidate_labels=labels, truncation=True)
            for label, score in zip(result['labels'], result['scores']):
                results.append({'xml_id': xml_id, 'sentence': sentence, 'SDG': label.split(":")[0], 'Score': score})

    # Convert results to DataFrame for easy inspection
    results_df = pd.DataFrame(results)
    # Calculate dynamic thresholds based on average score plus margin
    for label_name in label_names:
        avg_score = results_df[results_df['SDG'] == label_name]['Score'].mean()
        thresholds[label_name] = avg_score + 0.1  # Adding margin for threshold

    return results_df, thresholds



# Classify the DataFrame
claims_df, thresholds = classify_claims(df, batch_size=4)

# Print the DataFrame and thresholds
print("Claims DataFrame:", claims_df)
print("Thresholds:", thresholds)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords, words
from nltk.stem import WordNetLemmatizer
from scipy.sparse import csr_matrix
import re
import nltk

# Ensure NLTK resources are downloaded
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('words')

# Initialize stopwords, lemmatizer, and English word list
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
english_words = set(words.words())

# Preprocess function for text: remove punctuation, non-English words, stopwords, and lemmatize
def preprocess_text(text):
    # Remove punctuation, including hyphens, and lowercase
    text = re.sub(r"[^\w\s]", " ", text.lower())  # Removes punctuation, keeping only words and whitespace
    words_list = text.split()
    # Filter out stop words and non-English words, and lemmatize
    processed_words = [
        lemmatizer.lemmatize(word) for word in words_list 
        if word in english_words and word not in stop_words
    ]
    return ' '.join(processed_words)

# Load and prepare data (assuming 'df' is already loaded with 'marketing', 'pl1', 'pl2', 'brand', 'date')
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].apply(lambda x: "2023" if pd.Timestamp("2022-08-01") <= x <= pd.Timestamp("2023-08-31") 
                               else ("2024" if pd.Timestamp("2023-08-01") <= x <= pd.Timestamp("2024-08-31") else None))
df = df.dropna(subset=['year'])  # Drop rows outside the defined date ranges
df['year_pl1'] = df['year'] + '_' + df['pl1']
df['year_pl2'] = df['year'] + '_' + df['pl2']
df['year_brand'] = df['year'] + '_' + df['brands']

# Apply preprocessing to the marketing column before vectorizing
df['processed_marketing'] = df['marketing'].apply(preprocess_text)

# Initialize CountVectorizer and transform text to sparse matrix
vectorizer = CountVectorizer(ngram_range=(1, 2))  # Unigrams and bigrams
counts_sparse = vectorizer.fit_transform(df['processed_marketing'])

# Create sparse DataFrame with words and categories
vectorized_df = pd.DataFrame.sparse.from_spmatrix(
    csr_matrix(counts_sparse), columns=vectorizer.get_feature_names_out()
)
vectorized_df['year_pl1'] = df['year_pl1'].values
vectorized_df['year_pl2'] = df['year_pl2'].values
vectorized_df['year_brand'] = df['year_brand'].values

# Function to compute counts for each category
def compute_counts_by_category(vectorized_df, category_col):
    # Group by category-year combination and sum counts
    group = vectorized_df.groupby(category_col).sum()
    category_counts = group.stack().reset_index()
    category_counts.columns = [category_col, 'word_phrase', 'count']

    # Split year from category label and separate counts for 2023 and 2024
    category_counts['year'] = category_counts[category_col].str[:4]
    category_counts[category_col] = category_counts[category_col].str[5:]
    category_counts = category_counts.pivot(index=['word_phrase', category_col], columns='year', values='count').reset_index()
    category_counts.columns.name = None
    category_counts = category_counts.rename(columns={'2023': 'count_2023', '2024': 'count_2024'})

    # Add an ID column
    category_counts.insert(0, 'id', range(1, len(category_counts) + 1))
    return category_counts



```

```python
# Apply to each category
pl1_counts = compute_counts_by_category(vectorized_df, 'year_pl1')


# Display the results
print("Counts by 'pl1':")
print(pl1_counts.head())
```

```python
import numpy as np

def normalize_counts_by_product_count(pl1_counts, df):
    # Extract 'pl1' from 'year_pl1' in pl1_counts (since 'year_pl1' only contains the 'pl1' value)
    pl1_counts = pl1_counts.assign(pl1=pl1_counts['year_pl1'])

    # Calculate the number of products for each 'pl1' category and year in the original df
    product_counts = (
        df.groupby('pl1')
          .agg(num_products_2023=('year', lambda x: (x == '2023').sum()),
               num_products_2024=('year', lambda x: (x == '2024').sum()))
          .reset_index()
    )
    
    # Merge with pl1_counts to add product counts for normalization
    pl1_normalized_counts = pl1_counts.merge(
        product_counts,
        how='left',
        left_on=['pl1'],
        right_on=['pl1']
    )
    
    # Fill NaN values in count columns with 0
    pl1_normalized_counts['count_2023'] = pl1_normalized_counts['count_2023'].fillna(0)
    pl1_normalized_counts['count_2024'] = pl1_normalized_counts['count_2024'].fillna(0)
    
    # Normalize counts for each year
    pl1_normalized_counts['normalized_count_2023'] = (
        pd.to_numeric(pl1_normalized_counts['count_2023'], errors='coerce').fillna(0) / 
        pl1_normalized_counts['num_products_2023'].replace(0, 1)
    )
    pl1_normalized_counts['normalized_count_2024'] = (
        pd.to_numeric(pl1_normalized_counts['count_2024'], errors='coerce').fillna(0) / 
        pl1_normalized_counts['num_products_2024'].replace(0, 1)
    )
    
    # Calculate percentage increase for raw counts and normalized counts
    pl1_normalized_counts['percentage_increase_count'] = (
        (pd.to_numeric(pl1_normalized_counts['count_2024'], errors='coerce') - pd.to_numeric(pl1_normalized_counts['count_2023'], errors='coerce')) /
        pl1_normalized_counts['count_2023'].replace(0, np.nan)
    ).fillna(0) * 100
    pl1_normalized_counts['percentage_increase_normalized'] = (
        (pl1_normalized_counts['normalized_count_2024'] - pl1_normalized_counts['normalized_count_2023']) / 
        pl1_normalized_counts['normalized_count_2023'].replace(0, np.nan)
    ).fillna(0) * 100
    
    # Drop auxiliary columns
    pl1_normalized_counts = pl1_normalized_counts.drop(
        columns=['year', 'year_pl1']
    )
    
    # Drop duplicates
    pl1_normalized_counts = pl1_normalized_counts.drop_duplicates()
    
    return pl1_normalized_counts




norm_pl1_counts = normalize_counts_by_product_count(pl1_counts,df)
```
