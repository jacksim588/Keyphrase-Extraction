```python
import pandas as pd
```

```python
df_old = pd.read_csv(r'Extracted Phrases\Rake NLTK\Brandbank Bulk Phrase Extraction.csv')
df_old
```

```python
df_new = pd.read_csv(r'Extracted Phrases\Rake NLTK\Brandbank Latest Phrase Extraction.csv')
df_new
```

Find New Key Phrases

```python
list(set(df_new['Phrase']) - set(df_old['Phrase']))
```

Compare existing Phrases

```python
df_combined = df_new.merge(df_old, on='Phrase', how='left')
df_combined
```

```python
df_combined = df_combined.rename(columns={'Perc_x': 'Score_Frequency_x', 
                                            'Perc_y': 'Score_Frequency_y'},
                                            )
```

```python
def get_perc_increase(row):
    return ((row['Score_Frequency_x']-row['Score_Frequency_y'])/row['Score_Frequency_y'])*100

def get_score_increase(row):
    return ((row['Score_x']-row['Score_y'])/row['Score_y'])*100

def get_count_frequency_increase_x(row,sum):
    return (row['Count_x']/sum*100)

def get_count_frequency_increase_y(row,sum):
    return (row['Count_y']/sum*100)

def get_count_increase(row):
    if row['Count_Frequency_y'] !=0:
        return ((row['Count_Frequency_x']-row['Count_Frequency_y'])/row['Count_Frequency_y'])*100
    else:
        return 0
```

```python
sum = df_combined['Count_x'].sum()
df_combined['Count_Frequency_x'] = df_combined.apply (lambda row: get_count_frequency_increase_x(row,sum), axis=1)
df_combined['Count_Frequency_y'] = df_combined.apply (lambda row: get_count_frequency_increase_y(row,sum), axis=1)
df_combined['Count_Frequency_Increase'] = df_combined.apply (lambda row: get_count_increase(row), axis=1)
```

```python
df_combined['Score_Frequency_increase'] = df_combined.apply (lambda row: get_perc_increase(row), axis=1)
```

```python
df_combined
```

```python
df_combined['Score_increase'] = df_combined.apply (lambda row: get_score_increase(row), axis=1)
```
